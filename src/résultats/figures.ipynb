```python
# Template
fig1 = plt.figure()
x1 = [1,2,3]; y1 = [1,2,3]
x2 = [1,2,3]; y2 = [6,1,3]

plt.plot(x1, y1, label="Fonction 1")
plt.plot(x2, y2, label="Fonction 2")
plt.legend(loc = "upper left")
plt.show()
```
