```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()
pos = nx.spring_layout(G, seed=3113794652)  # positions for all nodes

# nodes
options = {"edgecolors": "tab:gray", "node_size": 800, "alpha": 0.9}
nx.draw_networkx_nodes(G, pos, nodelist=[0, 1, 2, 3], node_color="black", **options)
nx.draw_networkx_nodes(G, pos, nodelist=[4, 5, 6, 7], node_color="tab:blue", **options)

# edges
nx.draw_networkx_edges(G, pos, width=1.0, alpha=0.5)
nx.draw_networkx_edges(
    G,
    pos,
    edgelist=[(0, 1), (1, 2), (2, 3), (3, 0)],
    width=8,
    alpha=0.5,
    edge_color="tab:red",
)
nx.draw_networkx_edges(
    G,
    pos,
    edgelist=[(4, 5), (5, 6), (6, 7), (7, 4)],
    width=8,
    alpha=0.5,
    edge_color="tab:blue",
)


# some math labels
labels = {}
labels[0] = r"$a$"
labels[1] = r"$b$"
labels[2] = r"$c$"
labels[3] = r"$d$"
labels[4] = r"$\alpha$"
labels[5] = r"$\beta$"
labels[6] = r"$\gamma$"
labels[7] = r"$\delta$"
nx.draw_networkx_labels(G, pos, labels, font_size=22, font_color="whitesmoke")

plt.tight_layout()
plt.axis("off")
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx
def Merge(dict1, dict2):
    return(dict2.update(dict1))

G = nx.Graph()
G.add_edge(1, 2)
G.add_edge(2, 3)
G.add_edge(3, 4)
G.add_edge(4, 5)
G.add_edge(5, 6)



# Draw the nodes
pos_maps = {1: (0, 0), 2: (1, 0), 3: (2, 0), 4: (3, 0), 5: (4, 0)}
pos_tensors = {6: (0, 1), 7: (1, 1), 8: (0, 1), 9: (1, 1), 10: (4, 1)}
options = {"edgecolors": "tab:gray", "node_size": 800, "alpha": 1}
nx.draw_networkx_nodes(G, pos_maps, nodelist=[1, 2, 3, 4, 5], node_color="tab:red", **options)
nx.draw_networkx_nodes(G, pos_tensors, nodelist=[6, 7, 8, 9, 10], node_color="tab:blue", **options)

# Draw the edges
merged = Merge(pos_maps, pos_tensors)
nx.draw_networkx_edges(
    G,
    pos_maps | pos_tensors,
    width=8,
    alpha=0.5,
    edge_color="tab:red",
)

# Draw the labels
# some math labels
labels = {}
labels[0] = r"$a$"
labels[1] = r"$b$"
labels[2] = r"$c$"
labels[3] = r"$d$"
labels[4] = r"$\alpha$"
nx.draw_networkx_labels(G, pos_tensors, labels, font_size=22, font_color="whitesmoke")

plt.tight_layout()
plt.axis("off")
plt.show()
```
