```python
import matplotlib.pyplot as plt
import networkx as nx
def Merge(dict1, dict2):
    return(dict2.update(dict1))

G = nx.Graph(k=0.1)
G.add_edge(3, 0)
G.add_edge(1, 2)
G.add_edge(2, 3)
G.add_edge(3, 4)
G.add_edge(4, 5)

G.add_edge(1, 6)
G.add_edge(2, 7)
G.add_edge(3, 8)
G.add_edge(4, 9)
G.add_edge(5, 10)


# Draw the nodes
pos_empty = {0: (2, 1.01)}
pos_maps = {1: (0, 1), 2: (1, 1), 3: (2, 1), 4: (3, 1), 5: (4, 1)}
pos_tensors = {6: (0, 0.99), 7: (1, 0.99), 8: (2, 0.99), 9: (3, 0.99), 10: (4, 0.99)}
options = {
    "node_size": 1500,
    "edgecolors": "black",
    "linewidths": 3
}
nx.draw_networkx_nodes(G, pos_maps, nodelist=[1, 2, 3, 4, 5], node_color="white", **options)
nx.draw_networkx_nodes(G, pos_tensors, nodelist=[6, 7, 8, 9, 10], node_color="lightgray", **options)

# Draw the edges
merged = Merge(pos_maps, pos_tensors)
nx.draw_networkx_edges(
    G,
    pos_empty | pos_maps | pos_tensors,
    width=3,
)

# Draw the labels
# some math labels
labels = {}
labels[1] = r"$t_1$"
labels[2] = r"$t_2$"
labels[3] = r"$t_3$"
labels[4] = r"$\dots$"
labels[5] = r"$t_d$"
labels[6] = r"$\phi_{ml}(x_1)$"
labels[7] = r"$\phi_{ml}(x_2)$"
labels[8] = r"$\phi_{ml}(x_3)$"
labels[9] = r"$\dots$"
labels[10] = r"$\phi_{ml}(x_d)$"

nx.draw_networkx_labels(G, pos_maps | pos_tensors, labels, font_size=10)

edge_labels={
             (3, 0): r"$h$",
             (1, 2): r"$b_1$",
             (2, 3): r"$b_2$",
             (3, 4): r"$b_3$",
             (4, 5): r"$b_d$",
             (1, 6): r"$i_1$",
             (2, 7): r"$i_2$",
             (3, 8): r"$i_3$",
             (4, 9): r"$\dots$",
             (5, 10): r"$i_d$",
             }
nx.draw_networkx_edge_labels(
    G, pos_empty | pos_maps | pos_tensors,
    edge_labels=edge_labels,
    font_color='black',
    rotate=False
)

ax = plt.gca()
ax.margins(x=0.05, y=0.20)
plt.axis("off")
plt.show()
```
